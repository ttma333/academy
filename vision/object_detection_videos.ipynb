Base64 인코딩

https://cloud.google.com/video-intelligence/docs/base64?hl=ko

```python
def play(filename):
    html = ''
    video = open(filename,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1000 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

```python
# path at the end of the detection process "Results saved to yolov5/runs/detect/exp" + name from filename
play('/content/yolov5/runs/detect/exp/watch_v_NcaGFp76BTY.mp4') # uable to play a video in a notebook 
```
